```python
import matplotlib.pyplot as plt
import cv2

# Assuming 'image_rgb' is the image loaded in RGB format using OpenCV
plt.figure(figsize=(6, 10))

# Splitting the image into its Blue, Green, and Red components
blue, green, red = cv2.split(image_rgb)

# Plotting the original image
plt.subplot(2, 2, 1)  # 1st plot in the 2x2 grid
plt.imshow(cv2.cvtColor(image_rgb, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for proper display
plt.title("Original Image")
plt.axis('off')

# Plotting the Blue channel
plt.subplot(2, 2, 2)  # 2nd plot
plt.imshow(blue, cmap='gray')  # Display the blue channel in grayscale
plt.title("Blue Channel")
plt.axis('off')

# Plotting the Green channel
plt.subplot(2, 2, 3)  # 3rd plot
plt.imshow(green, cmap='gray')
plt.title("Green Channel")
plt.axis('off')

# Plotting the Red channel
plt.subplot(2, 2, 4)  # 4th plot
plt.imshow(red, cmap='gray')
plt.title("Red Channel")
plt.axis('off')

plt.tight_layout()  # Adjust spacing between plots
plt.show()
```
