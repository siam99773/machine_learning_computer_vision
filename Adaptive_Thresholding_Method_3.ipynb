```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = '/content/berry-1.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

blurred_image = cv2.GaussianBlur(image, (5, 5), 0)


adaptive_thresh_gaussian = cv2.adaptiveThreshold(blurred_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                                 cv2.THRESH_BINARY, 11, 2)

kernel = np.ones((3, 3), np.uint8)
cleaned_image = cv2.morphologyEx(adaptive_thresh_gaussian, cv2.MORPH_OPEN, kernel)


plt.figure(figsize=(6, 6))
plt.imshow(cleaned_image, cmap='gray')
plt.title('Berry Separated from Background with Adaptive Thresholding')
plt.show()


cv2.imwrite('/content/berry_separated.jpg', cleaned_image)
```

Gaussian Blur: This reduces image noise, making the thresholding process more effective by smoothing out small imperfections and details that aren't relevant.

Adaptive Gaussian Thresholding: It adapts to different lighting conditions in the image, calculating thresholds for small regions. This makes it ideal for separating the berry from the unevenly lit background.

Morphological Operations: Using the "Opening" operation helps remove small noise from the background, ensuring the berry is cleanly separated from unwanted artifacts.
