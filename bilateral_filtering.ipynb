**Bilateral Filtering:
Bilateral filtering is a non-linear, edge-preserving, and noise-reducing smoothing filter commonly used in image processing. Unlike Gaussian blur, which simply averages pixel values, bilateral filtering considers both the spatial distance and the intensity difference of nearby pixels, thus maintaining sharp edges while reducing noise.
Bilateral filtering works by averaging nearby pixel values
Spatial proximity: How close the pixel is to the one being processed.
Intensity similarity: How similar the intensity (color or brightness) of the nearby pixel is to the one being processed.**


In OpenCV, the bilateralFilter() function is used to apply bilateral filtering on an image. It takes several parameters, two of the most important ones are:

sigmaColor:

It represents the filterâ€™s range sensitivity to pixel intensity (color difference).
A higher value of sigmaColor means that pixels with larger intensity differences will be considered as neighbors and included in the average, even if they have a significant intensity difference from the center pixel.
If sigmaColor is small, only pixels with similar colors or intensity levels will have an impact on the center pixel, thus preserving strong edges better.
sigmaSpace:

It defines how far the spatial influence of nearby pixels is.
A larger sigmaSpace means that pixels farther away from the center pixel (in terms of spatial distance) will influence the filtering process.
If sigmaSpace is small, only nearby pixels will influence the result, and as it increases, more distant pixels will have a say in the weighted average, though their effect will still depend on the value of sigmaColor.

```python
import cv2
import matplotlib.pyplot as plt


image = cv2.imread('/content/orange-1.jpg')


filtered_image = cv2.bilateralFilter(image, d=15, sigmaColor=75, sigmaSpace=75)


filtered_image_rgb = cv2.cvtColor(filtered_image, cv2.COLOR_BGR2RGB)


plt.imshow(filtered_image_rgb)
plt.title('Bilateral Filter')
plt.axis('off')  # Hide axes
plt.show()
```
