```python

import cv2
import matplotlib.pyplot as plt
import numpy as np


image_path = '/content/berry-1.jpg'


image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

_, thresh_binary = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
_, thresh_binary_inv = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
_, thresh_trunc = cv2.threshold(image, 127, 255, cv2.THRESH_TRUNC)
_, thresh_tozero = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO)
_, thresh_tozero_inv = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO_INV)


adaptive_thresh_mean = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
adaptive_thresh_gaussian = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

# Plotting the results
plt.figure(figsize=(10, 8))

plt.subplot(2, 3, 1)
plt.imshow(thresh_binary, cmap='gray')
plt.title('Binary Threshold')

plt.subplot(2, 3, 2)
plt.imshow(thresh_binary_inv, cmap='gray')
plt.title('Binary Inverse Threshold')

plt.subplot(2, 3, 3)
plt.imshow(thresh_trunc, cmap='gray')
plt.title('Trunc Threshold')

plt.subplot(2, 3, 4)
plt.imshow(thresh_tozero, cmap='gray')
plt.title('To Zero Threshold')

plt.subplot(2, 3, 5)
plt.imshow(thresh_tozero_inv, cmap='gray')
plt.title('To Zero Inverse Threshold')

plt.subplot(2, 3, 6)
plt.imshow(adaptive_thresh_mean, cmap='gray')
plt.title('Adaptive Mean Threshold')

plt.tight_layout()
plt.show()


plt.figure(figsize=(6, 6))
plt.imshow(adaptive_thresh_gaussian, cmap='gray')
plt.title('Adaptive Gaussian Threshold')
plt.show()
```

**Contribution of Image Thresholding in Computer Vision:
Image Segmentation: Thresholding helps separate objects from the background, critical in applications like medical imaging (tumor detection) and industrial inspection (defect detection).

Feature Extraction: It simplifies image analysis by highlighting contours and regions of interest, aiding tasks like shape analysis, object recognition, and blob detection.

Optical Character Recognition (OCR): Thresholding isolates text from the background, improving text recognition accuracy in scanned documents.

Edge Detection: It emphasizes sharp edges in an image, aiding in tasks like boundary detection and feature enhancement.
