```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image in grayscale
image = cv2.imread('/content/panda.jpeg', cv2.IMREAD_GRAYSCALE)


sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Sobel in x direction
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Sobel in y direction


sobel_x = cv2.convertScaleAbs(sobel_x)
sobel_y = cv2.convertScaleAbs(sobel_y)


sobel_combined_weighted = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)


sobel_combined_bitwise_or = cv2.bitwise_or(sobel_x, sobel_y)

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1), plt.imshow(sobel_x, cmap='gray'), plt.title('Sobel X')
plt.subplot(1, 3, 2), plt.imshow(sobel_y, cmap='gray'), plt.title('Sobel Y')
plt.subplot(1, 3, 3), plt.imshow(sobel_combined_weighted, cmap='gray'), plt.title('Combined (Weighted)')
plt.show()


plt.imshow(sobel_combined_bitwise_or, cmap='gray')
plt.title('Combined (Bitwise OR)')
plt.show()

# Bonus Explanation:
"""
The bitwise_or function performs a logical OR operation between the pixel values of two images.
This can work for combining Sobel filtered images, but it is generally not the preferred method because it does not
take into account the magnitude of the gradients in both directions.

The recommended approach is to use cv2.addWeighted(), which allows you to blend the horizontal and vertical gradients
in a controlled manner by weighting the contributions of each. This method better represents the intensity of edges in the image.

In summary, while bitwise_or can be used, cv2.addWeighted() is the preferred method for combining Sobel filtered images
as it provides more visually accurate results.
"""
```
